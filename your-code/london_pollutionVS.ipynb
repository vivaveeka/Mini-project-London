```python
import pandas as pd
emissions_summary = pd.read_excel('../../Mini-project-London/data/data-pollution/LAEI2016_Emissions_Summary-NOxPMCO2_modified.xlsx', sheet_name='MASTER')
emissions_summary
```

```python
education = pd.read_excel('../../Mini-project-London/data/data-education-modified/education-averages.xlsx', skiprows = 1, nrows = 35)
education = education.drop(labels=0)
education = education.drop(labels=34)
# df = df.rename(columns={'oldName1': 'newName1', 'oldName2': 'newName2'})
education = education.rename(columns = {"Area":"BOROUGHS"})
education
```

```python
new = pd.merge(emissions_summary, education, on="BOROUGHS", how="left")
new
```

```python
#pandas.DataFrame.corrwith() - run correlations between columns in different data frames
# Corr() seems to work fine with merged data frame

correlation = new.corr(method ='pearson')
correlation
```
